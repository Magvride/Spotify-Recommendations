## 📌 Visão Geral do Projeto

### Problem Statement - Motivação
Este projeto busca desenvolver um sistema de recomendação de músicas baseado em características de áudio (Content-Based) e outras variáveis. O objetivo é sugerir músicas com sonoridade similar a partir de uma faixa escolhida, sem depender do histórico de ouvintes. Esse problema é relevante pois ajuda a entender como algoritmos podem identificar padrões em dados complexos, além de ser uma aplicação prática e dinâmica em Ciência de Dados.

### Objetivos
- **Principal:** Construir um sistema de recomendação de músicas que, dado o nome de uma faixa, sugira 5 músicas similares com base em suas audio features.
- **Secundários:**
  - Explorar e compreender as distribuições das principais audio features do dataset do Spotify.

  - Aplicar técnicas de normalização e pré-processamento de dados.
  
  - Implementar e comparar métricas de similaridade.

  - Avaliar qualitativamente a coerência das recomendações.
